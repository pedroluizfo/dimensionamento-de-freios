```python
#importando bibliotecas.
import pandas as pd
import math as m
import os
#verificando e criando a pasta "pasta".
if os.path.exists('./pasta'):
    print('Já existe')
else: 
    os.mkdir('./pasta')

#desingn do dataframe
Df = pd.DataFrame(columns=['Distância','Tempo','Pressão Cil. Mestre [Pa]','Coeficiente de atrito','Area da Pastilha Furada [m²]','Raio do furo (mm)','Aceleração','Força Pinça [N]','Força Fricção [N]' ,'Torque de frenagem [N/m]','Area Max Furada [m²]', 'Num. furos','ForcaCaliperF',])
#criando repetição das entradas.
c = ''
while c != 'N':  

    vf = 19.44
    vi = 22.22
    d = float(input('Distância: '))
    print(f'Distância: {d}m')

    aceleração = (((vf*2)-(vi*2))/(2*d))
    tempo = vi/(aceleração*-1)
    print(f'tempo: {tempo}s')
    RaioEf = (99.875)/1000 
    Pressure = float(input('Pressão Cilindro Mestre [Pa]: '))
    print(f'Pressão do cilindro mestre: {Pressure}')

    AreaCaliper = float(706.86)*(10**(-6)) #não condiz com a realidade (5mm de raio)
    AreaPastilha = float(3521.856)*(10**(-6)) #não condiz com a realidade eu acho
    ForcaCaliper = 2*((Pressure*AreaCaliper)/AreaPastilha)
    print(f'Força Pinça: {ForcaCaliper}')

    mi = float(input('Coeficiente de atrito: '))
    ForcaFriccao = ForcaCaliper*mi
    print(f'Força Fricção: {ForcaFriccao}')

    TorqueFrenagem = ForcaFriccao*RaioEf
    print(f'Torque de frenagem: {TorqueFrenagem}')

    AreaPastilhaF = float(input('Area da Pastilha Furada [m**2]: '))
    print(f'Área da pastilha furada: {AreaPastilhaF}')
    AreaMaximaFurada = AreaPastilha-AreaPastilhaF
    print(f'Area Máxima que pode ser furada: {AreaMaximaFurada}')

    RaioFuro = float(input('Raio do furo: ')) 
    NumFuros = AreaMaximaFurada//(m.pi*((RaioFuro/1000)**2))
    print(f'Numero de furos que podem ser feitos: {NumFuros}') 
   

    ForcaCaliperF =  2*((Pressure*AreaCaliper)/AreaPastilhaF)
    ForcaFriccaoF = ForcaCaliperF*mi  
    TorqueFrenagemF = ForcaFriccaoF*RaioEf

    print(f'Força caliper "furado"{ForcaCaliperF}')
    print(f'força friccção{ForcaFriccaoF}')
    print(f'força frenagem {TorqueFrenagemF}')


    c = str(input('Continuar? ')).capitalize()
    Df = Df.append(pd.Series( {'Distância':d,'Tempo':tempo,'Pressão Cil. Mestre [Pa]': Pressure ,'Coeficiente de atrito': mi,'Area da Pastilha Furada [m²]': AreaPastilhaF ,'Raio do furo (mm)': RaioFuro,'Aceleração':aceleração ,'Força Pinça [N]': ForcaCaliper,'Força Fricção [N]': ForcaFriccao,'Torque de frenagem [N/m]':TorqueFrenagem,'Area Max Furada [m²]': AreaMaximaFurada, 'Num. furos': NumFuros,'ForcaCaliperF':ForcaCaliperF,'ForcaFriccaoF':ForcaFriccaoF,'TorqueFrenagemF':TorqueFrenagemF}), ignore_index=True)
print(Df)
continuar = str(input('Deseja salvar o arquivo?[S] ou [N] ')).capitalize()
if continuar == 'S':
    pasta='./pasta'
    cont = 1
    for diretorio, subpastas, arquivos in os.walk(pasta):
        for arquivo in arquivos:
            cont = cont+1
            print(os.path.join(diretorio, arquivo))
            print(cont)
    writer = pd.ExcelWriter(f'./pasta/Dados de frenagem teste {cont}.xlsx')
    Df.to_excel(writer)
    writer.save()
    print("DataFrame is exported successfully to 'Dados de frenagem teste' Excel File.")

    df_novo= pd.read_excel(f'pasta\dados de frenagem teste {cont}.xlsx')
else:
    print('O programa foi encerrado sem salvar os dados!')
```
