```python

#importando bibliotecas.
import pandas as pd
import math as m
import os
#verificando e criando a pasta "pasta".
if os.path.exists('./pasta'):
    print('Já existe')
else: 
    os.mkdir('./pasta')

#Criando listas vazias para salvar as entradas []
l1,l2,l3,l4,l5,l6,l7,l8,l9,l10,l11,l12,l13,l14,l15 = [],[],[],[],[],[],[],[],[],[],[],[],[],[],[]

#criando repetição das entradas.
c = ''
while c != 'N':  

    vf = 19.44
    vi = 22.22
    massa = 850
    RaioEf = (99.875)/1000
    AreaCaliper = float(706.86)*(10**(-6)) #não condiz com a realidade (5mm de raio)
    AreaPastilha = float(3521.856)*(10**(-6)) #não condiz com a realidade eu acho
    RaioRoda = 0.1155
    
    d = float(input('Distância [m]: '))
    print(f'Distância: {d} m')
    aceleração = (((vf**2)-(vi**2))/(2*d))
    print(f'Aceleração: {(aceleração * -1):.4f} m/s²')
    tempo = vi/(aceleração*-1)
    print(f'Tempo: {tempo:.4f} s')
    
    ForcaTotal = massa * aceleração
    TorqueTotal = (ForcaTotal/4) * RaioRoda
    
    Pressure = float(input('Pressão Cilindro Mestre [Pa]: '))
    print(f'Pressão do cilindro mestre: {Pressure:.4f} Pa')

    ForcaCaliper = 2*((Pressure*AreaCaliper)/AreaPastilha)
    print(f'Força Pinça: {ForcaCaliper:.4f} N')

    mi = float(input('Coeficiente de atrito: '))
    ForcaFriccao = ForcaCaliper*mi
    print(f'Força Fricção: {ForcaFriccao:.4f} N')

    TorqueFrenagem = ForcaFriccao*RaioEf
    print(f'Torque de frenagem: {TorqueFrenagem:.4f} N/m')

    if ForcaTotal > ForcaCaliper:
        print('Refazer parametros, Força insuficiente para frenagem')
    else: 
        print('Força suficientes para frenagem')
    
    if TorqueTotal > TorqueFrenagem:
        print('Refazer parametros, Torque insuficiente para frenagem')
    else: 
        print('Torque suficientes para frenagem')

    AreaPastilhaF = float(input('Nova Area da Pastilha Furada [m²]: '))
    print(f'Nova Area da pastilha furada: {AreaPastilhaF:.5f} m²')
    AreaMaximaFurada = AreaPastilha-AreaPastilhaF
    print(f'Area Máxima que pode ser furada [m²]: {AreaMaximaFurada:.6f}')

    RaioFuro = float(input('Raio do furo [mm]: ')) 
    NumFuros = AreaMaximaFurada//(m.pi*((RaioFuro/1000)**2))
    print(f'Numero de furos que podem ser feitos: {NumFuros:.4f}') 
   
    ForcaCaliperF =  2*((Pressure*AreaCaliper)/AreaPastilhaF)
    ForcaFriccaoF = ForcaCaliperF*mi  
    TorqueFrenagemF = ForcaFriccaoF*RaioEf

    print(f'Nova Força caliper {ForcaCaliperF:.4f} N')
    print(f'Nova Força friccção {ForcaFriccaoF:.4f} N')
    print(f'Nova Torque frenagem {TorqueFrenagemF:.4f} N/m')
    
    if ForcaTotal > ForcaCaliperF:
        print('Refazer parametros, nova força insuficiente para frenagem')
    else: 
        print('Nova força suficiente para frenagem')
    
    if TorqueTotal > TorqueFrenagemF:
        print('Refazer parametros, novo torque insuficiente para frenagem')
    else: 
        print('Novo torque suficiente para frenagem')

    
    c = str(input('Continuar? ')).capitalize()
#salvando cada input em sua respectiva lista
    l1.append(d)
    l2.append(tempo)
    l3.append(Pressure)
    l4.append(mi)
    l5.append(AreaPastilhaF)
    l6.append(RaioFuro)
    l7.append(aceleração)
    l8.append(ForcaCaliper)
    l9.append(ForcaFriccao)
    l10.append(TorqueFrenagem)
    l11.append(AreaMaximaFurada)
    l12.append(NumFuros)
    l13.append(ForcaCaliperF)
    l14.append(ForcaFriccaoF)
    l15.append(TorqueFrenagemF)

Df = pd.DataFrame({'Distância':l1,'Tempo':l2,'Pressão Cil. Mestre [Pa]': l3 ,'Coeficiente de atrito': l4,'Area da Pastilha Furada [m²]': l5 ,'Raio do furo (mm)': l6,'Aceleração':l7 ,'Força Pinça [N]': l8,'Força Fricção [N]': l9,'Torque de frenagem [N/m]':l10,'Area Max Furada [m²]': l11, 'Num. furos': l12,'ForcaCaliperF':l13,'ForcaFriccaoF':l14,'TorqueFrenagemF':l15})
pd.concat([Df],axis = 1)


continuar = str(input('Deseja salvar o arquivo? [S] ou [N] ')).capitalize()
if continuar == 'S':
    pasta='./pasta'
    cont = 1
    for diretorio, subpastas, arquivos in os.walk(pasta):
        for arquivo in arquivos:
            cont = cont+1
            print(os.path.join(diretorio, arquivo))
            print(cont)
    writer = pd.ExcelWriter(f'./pasta/Dados de frenagem teste {cont}.xlsx')
    Df.to_excel(writer)
    writer.save()
    print("DataFrame is exported successfully to 'Dados de frenagem teste' Excel File.")

else:
    print('O programa foi encerrado sem salvar os dados!')

```
